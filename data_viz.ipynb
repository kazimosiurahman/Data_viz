```python
#moving ahead with subplots
plt.subplot(1,2,1)
plt.title("1st chart (Sin)")
plt.plot(x,y,'b')

plt.subplot(1,2,2)
plt.title("2nd Chart (Cos Inveresed)")
plt.plot(y,x,'r')
plt.tight_layout()
```

```python
#object oriented way single plot

#figure is an object created. the size parameters can be adjusted
# in figure there is a method call axes which acts like aspect ration
fig = plt.figure(figsize=(10,8))
ax=fig.add_axes([0.1,0.1,0.9,0.5])
ax.plot(x,y,'b')
ax.plot(x,y,x,z)
```

```python
#Object Oriented way of Overlapping plot
fig = plt.figure()

ax1=fig.add_axes([0.1,0.1,0.9,0.9])
ax2=fig.add_axes([0.4,0.7,0.25,0.25])
ax1.plot(x,y,'m')
ax2.plot(x,y,'b')
ax1.set_title("Bigger graph",color='black')
ax2.set_title("Smaller graph",color='black')
```

```python
#changing the axis range
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.9,0.9])
x = np.arange(0,5,0.01)
y = np.sin(x)
ax.plot(x,y)
ax.set_xlim([-5,0.5])
ax.set_ylim([-1,1.5])
```

```python
#till now the plottings have been done on a linear scale. When you have huge amount of data, then we need to get to a log scale
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.9,0.9])
ax.set_yscale('log')
ax.set_xticks([0,2.5,5])
ax.plot(x,x*x*x*x*x)
```

```python
#step plot where the connections are not smooth as earlier
plt.style.use('ggplot')
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.9,0.9])
x = np.arange(0.1,10,0.3)
y = np.sin(x)

ax.step(x,y,lw=3)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.9,0.9])
x = np.arange(1,20,0.2)
y = np.cos(x)
ax.fill_between(x,y,0.3*y,color='#FF9F00')
ax.set_xlim([1.5,11.0])
ax.set_ylim([-1,0.25])
# ax.plot(x,y)
```

```python
#Using your own custom grid
plt.style.use('default')
fig = plt.figure(figsize=(3,2))
ax = fig.add_axes([0.1,0.1,0.9,0.9])
ax.grid(color='b',ls = 'dashed')
ax.plot(x,y,'r')
```

```python
#Adding text annotations to a graph
plt.style.use('default')
fig = plt.figure(figsize=(3,2))
ax = fig.add_axes([0.1,0.1,0.9,0.9])
y = x*x
ax.grid(color='b',ls = 'dashed')
ax.plot(x,y,'r')
ax.text(8,200,"$y = x^2$",fontsize=15,color='black')
```

```python
tips=sns.load_dataset('tips')
titanic = sns.load_dataset('titanic')
```

```python
#count plot usually when one variable is categorical
titanic.sex.unique()
titanic['class'].unique()
```

```python
sns.countplot(x = 'class',data=titanic)
```

```python
#when you wish to add another dimension to the label sex let's say categorising them according to classes use hue
sns.countplot(x='sex',data=titanic,hue='class')
```

```python
#strip plot is like scatterplot however for categorical 
# sns.countplot(x='sex',data=tips,hue='day')
sns.stripplot(x='sex',y='total_bill',hue='day',data=tips,)
```

```python
#trying swarm plot on the same set 
#swarm plot is similar to strip plot without any overlapping
#comparing the above and below charts, male and femlae plots have kind of overlapped , whereas in swarm it has spread
sns.swarmplot(x='sex',y='total_bill',hue='day',data=tips,)
```

```python
#distribution of the total bill 
sns.violinplot(x='sex',y='total_bill',data=tips,)
```

```python
#bar plot - point estimates for categorical data
# box plot - distribution for each category
sns.barplot(x='sex',y='age',hue='class',data=titanic,order=['female','male'],estimator=np.std,palette='Set1')
```

```python
sns.boxplot(x='sex',y='age',hue='class',data=titanic,)
```
