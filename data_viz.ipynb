```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10,0.2)
y = np.sin(x)
z = np.cos(x)
```

```python
#to check the availability of various styles
plt.style.available
```

```python
#using the ggplot style the below function is used to set
plt.style.use('ggplot')
```

```python
plt.plot(x,y)
plt.plot(x,z)
plt.plot(y,z)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Different grpahical methods when plotted against each other")
```

```python
#drawing multiple plots in a file, 
#drawing multiple plots in subplots
#you can also plot the figures like before to overlap

plt.plot(x,y,x,z)
```

```python
#moving ahead with subplots
plt.subplot(1,2,1)
plt.title("1st chart (Sin)")
plt.plot(x,y,'b')

plt.subplot(1,2,2)
plt.title("2nd Chart (Cos Inveresed)")
plt.plot(y,x,'r')
plt.tight_layout()
```

```python
#object oriented way single plot

#figure is an object created. the size parameters can be adjusted
# in figure there is a method call axes which acts like aspect ration
fig = plt.figure(figsize=(10,8))
ax=fig.add_axes([0.1,0.1,0.9,0.5])
ax.plot(x,y,'b')
ax.plot(x,y,x,z)
```

```python
#Object Oriented way of Overlapping plot
fig = plt.figure()

ax1=fig.add_axes([0.1,0.1,0.9,0.9])
ax2=fig.add_axes([0.4,0.7,0.25,0.25])
ax1.plot(x,y,'m')
ax2.plot(x,y,'b')
ax1.set_title("Bigger graph",color='black')
ax2.set_title("Smaller graph",color='black')
```

```python
fig = plt.figure()
x = np.arange(0,10,1)
y = np.square(x)
ax1= fig.add_axes([0.1,0.1,0.9,0.9])
ax2 = fig.add_axes([0.1,0.3,0.25,0.25])
ax1.plot(x,y,color='b')
ax2.plot(x,y,color='r')
ax1.set_title("Bigger graph", color="black")
ax2.set_title("Smaller graph", color="black")
```
