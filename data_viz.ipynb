```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10,0.2)
y = np.sin(x)
z = np.cos(x)
```

```python
#to check the availability of various styles
plt.style.available
```

```python
#using the ggplot style the below function is used to set
plt.style.use('ggplot')
```

```python
plt.plot(x,y)
plt.plot(x,z)
plt.plot(y,z)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Different grpahical methods when plotted against each other")
```
