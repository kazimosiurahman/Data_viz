```python
#drawing multiple plots in a file, 
#drawing multiple plots in subplots
#you can also plot the figures like before to overlap

plt.plot(x,y,x,z)
```

```python
#moving ahead with subplots
plt.subplot(1,2,1)
plt.title("1st chart (Sin)")
plt.plot(x,y,'b')

plt.subplot(1,2,2)
plt.title("2nd Chart (Cos Inveresed)")
plt.plot(y,x,'r')
plt.tight_layout()
```

```python
#object oriented way single plot

#figure is an object created. the size parameters can be adjusted
# in figure there is a method call axes which acts like aspect ration
fig = plt.figure(figsize=(10,8))
ax=fig.add_axes([0.1,0.1,0.9,0.5])
ax.plot(x,y,'b')
ax.plot(x,y,x,z)
```

```python
#Object Oriented way of Overlapping plot
fig = plt.figure()

ax1=fig.add_axes([0.1,0.1,0.9,0.9])
ax2=fig.add_axes([0.4,0.7,0.25,0.25])
ax1.plot(x,y,'m')
ax2.plot(x,y,'b')
ax1.set_title("Bigger graph",color='black')
ax2.set_title("Smaller graph",color='black')
```

```python
fig = plt.figure()
x = np.arange(0,10,1)
y = np.square(x)
ax1= fig.add_axes([0.1,0.1,0.9,0.9])
ax2 = fig.add_axes([0.1,0.3,0.25,0.25])
ax1.plot(x,y,color='b')
ax2.plot(x,y,color='r')
ax1.set_title("Bigger graph", color="black")
ax2.set_title("Smaller graph", color="black")
```

```python
#object oriented way : subplotting
fig,ax=plt.subplots(3,3,figsize=(10,5))
# fig = plt.figure()
ax[0,1].plot(x,y,'b')
ax[0,1].set_xlabel("X")
ax[0,1].set_ylabel("Y")
ax[0,1].set_title("1st Chart",color='black')

# ax[1].plot(y,x,'r')
# ax[1].set_xlabel("X")
# ax[1].set_ylabel("Y")
# ax[1].set_title("2nd Chart",color='black')
plt.tight_layout()
```

```python
#displaying a bar chart, when the data is categorical . summation of all the category value
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.3,0.9])
ax.bar(x = ['Male','Female'] ,height = [200,300] ,width = 0.7 ) 
# plt.tight_layout()
#to design the same plot horizontally you use another method called barh
fig = plt.figure()
ax1 = fig.add_axes([0.1,0.1,0.9,0.3])
ax1.barh(y = ['Male','Female'] ,width = [200,300],height =0.6) 
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.9,0.9])

y = np.sin(x)
ax.plot(x,y,'r',label='Sine')
ax.plot(y,x,'b',label='Inverse of Sine')
# ax.plot(x,y+10,'b',label='Inverse of Sine')
ax.legend()
fig.savefig('./Sine_inverse.jpg')
```

```python
#changing the axis range
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.9,0.9])
x = np.arange(0,5,0.01)
y = np.sin(x)
ax.plot(x,y)
ax.set_xlim([-5,0.5])
ax.set_ylim([-1,1.5])
```

```python
#till now the plottings have been done on a linear scale. When you have huge amount of data, then we need to get to a log scale
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.9,0.9])
ax.set_yscale('log')
ax.plot(x,x*x*x*x*x)
```
